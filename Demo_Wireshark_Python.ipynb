```python
# Instalasi pyshark (pastikan SHARK_PATH sudah diatur)
!pip install pyshark

# Mengatur SHARK_PATH
import os
os.environ['SHARK_PATH'] = '/usr/bin/tshark'  # Sesuaikan dengan lokasi tshark yang ada di sistem Anda

# Menggunakan pyshark untuk menangkap paket dari antarmuka jaringan
import pyshark

# Menangkap paket dari antarmuka jaringan (misalnya, 'eth0')
capture = pyshark.LiveCapture(interface='eth0')

# Menampilkan informasi dasar tentang setiap paket yang ditangkap
for packet in capture.sniff_continuously(packet_count=10):
    print(packet)
```
