```python
import requests
```

```python
# Menggunakan JSONPlaceholder sebagai contoh
url = "https://jsonplaceholder.typicode.com/posts"

# Data yang akan dikirim dalam permintaan POST
payload = {
    "title": "foo",
    "body": "bar",
    "userId": 1
}

# Header untuk permintaan
headers = {
    "Content-Type": "application/json"
}

# Mengirim permintaan POST dan mendapatkan respons
response = requests.post(url, json=payload, headers=headers)

# Menampilkan respons
print("Status Code:", response.status_code)
print("Response Body:", response.json())

# Menguji input yang tidak valid
invalid_payload = {
    "title": "foo'; DROP TABLE posts; --",
    "body": "bar",
    "userId": 1
}

# Mengirim permintaan dengan input tidak valid
invalid_response = requests.post(url, json=invalid_payload, headers=headers)

# Menampilkan respons untuk input tidak valid
print("Status Code (Invalid):", invalid_response.status_code)
print("Response Body (Invalid):", invalid_response.json())
```

```python

# Menggunakan JSONPlaceholder sebagai contoh
url = "https://jsonplaceholder.typicode.com/posts"

# Data yang akan dikirim dalam permintaan POST
payload = {
    "title": "foo",
    "body": "bar",
    "userId": 1
}

# Header untuk permintaan
headers = {
    "Content-Type": "application/json"
}

# Mengirim permintaan POST dan mendapatkan respons
response = requests.post(url, json=payload, headers=headers)

# Menampilkan respons
print("Status Code:", response.status_code)
print("Response Body:", response.json())

# Menguji input yang tidak valid

# 1. SQL Injection
sql_injection_payload = {
    "title": "foo'; DROP TABLE posts; --",
    "body": "bar",
    "userId": 1
}

sql_injection_response = requests.post(url, json=sql_injection_payload, headers=headers)
print("\nSQL Injection Test")
print("Status Code:", sql_injection_response.status_code)
print("Response Body:", sql_injection_response.json())

# 2. Cross-Site Scripting (XSS)
xss_payload = {
    "title": "<script>alert('XSS');</script>",
    "body": "bar",
    "userId": 1
}

xss_response = requests.post(url, json=xss_payload, headers=headers)
print("\nXSS Test")
print("Status Code:", xss_response.status_code)
print("Response Body:", xss_response.json())

# 3. Payload Berbahaya Lainnya
# Contoh payload JSON yang besar untuk menguji batasan ukuran
large_payload = {
    "title": "A" * 10000,  # String panjang
    "body": "B" * 10000,
    "userId": 1
}

large_payload_response = requests.post(url, json=large_payload, headers=headers)
print("\nLarge Payload Test")
print("Status Code:", large_payload_response.status_code)
print("Response Body:", large_payload_response.json())
```
