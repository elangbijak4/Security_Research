```python
import requests
import json
```

```python
# URL endpoint untuk mengakses CVE
url = "https://cve.circl.lu/api/cve/CVE-2021-44228"

# Mengirim permintaan GET ke API
response = requests.get(url)

# Memeriksa apakah permintaan berhasil
if response.status_code == 200:
    cve_data = response.json()  # Mengonversi respons ke format JSON
    print(json.dumps(cve_data, indent=4))  # Mencetak data CVE dengan format yang rapi
else:
    print(f"Gagal mengambil data CVE. Status kode: {response.status_code}")
```

```python
# URL endpoint untuk mengakses CVE
url = "https://cve.circl.lu/api/cve/CVE-2021-44228"

# Mengirim permintaan GET ke API
response = requests.get(url)

# Memeriksa apakah permintaan berhasil
if response.status_code == 200:
    cve_data = response.json()  # Mengonversi respons ke format JSON

    # Menampilkan informasi penting dari CVE
    print(f"ID CVE: {cve_data.get('id')}")
    print(f"Deskripsi: {cve_data.get('summary')}")
    print(f"Tanggal Diumumkan: {cve_data.get('Published')}")
    print(f"Skor CVSS: {cve_data.get('cvss')}")
    print(f"Kritisitas: {cve_data.get('cvss-vector')}")

    # Menampilkan referensi yang terkait
    print("\nReferensi:")
    for reference in cve_data.get('references', []):
        print(f"- {reference}")
else:
    print(f"Gagal mengambil data CVE. Status kode: {response.status_code}")
```
