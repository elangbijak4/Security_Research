```python
from scapy.all import *

def packet_callback(packet):
    if packet.haslayer(TCP):
        # Deteksi paket TLS (port 443)
        if packet[TCP].dport == 443 or packet[TCP].sport == 443:
            print(f"[{packet.time}] TLS Packet from {packet[IP].src}:{packet[TCP].sport} to {packet[IP].dst}:{packet[TCP].dport}")
        # Deteksi paket HTTP (port 80)
        elif packet[TCP].dport == 80 or packet[TCP].sport == 80:
            print(f"[{packet.time}] HTTP Packet from {packet[IP].src}:{packet[TCP].sport} to {packet[IP].dst}:{packet[TCP].dport}")
        # Tambahkan informasi tambahan untuk paket TCP lainnya
        else:
            print(f"[{packet.time}] TCP Packet from {packet[IP].src}:{packet[TCP].sport} to {packet[IP].dst}:{packet[TCP].dport}")

    elif packet.haslayer(UDP):
        # Deteksi paket DNS (port 53)
        if packet[UDP].dport == 53 or packet[UDP].sport == 53:
            print(f"[{packet.time}] DNS Packet from {packet[IP].src}:{packet[UDP].sport} to {packet[IP].dst}:{packet[UDP].dport}")
        # Tambahkan informasi tambahan untuk paket UDP lainnya
        else:
            print(f"[{packet.time}] UDP Packet from {packet[IP].src}:{packet[UDP].sport} to {packet[IP].dst}:{packet[UDP].dport}")

    elif packet.haslayer(ICMP):
        # Deteksi paket ICMP (ping)
        print(f"[{packet.time}] ICMP Packet from {packet[IP].src} to {packet[IP].dst} - Type: {packet[ICMP].type} Code: {packet[ICMP].code}")

    elif packet.haslayer(ARP):
        # Deteksi paket ARP spoofing
        if packet[ARP].op == 2:
            print(f"[*] ARP Spoof Packet from {packet[ARP].psrc} claiming to be {packet[ARP].hwsrc}")
        else:
            print(f"[{packet.time}] ARP Packet - Operation: {packet[ARP].op} From: {packet[ARP].psrc} To: {packet[ARP].pdst}")

    else:
        # Deteksi paket lainnya
        print(f"[{packet.time}] Other Packet: {packet.summary()}")

# Start sniffing the network
print("Starting network sniffing...")
sniff(prn=packet_callback, store=0)
```
